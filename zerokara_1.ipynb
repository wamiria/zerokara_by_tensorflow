```
import tensorflow as tf
```

1.3.1 算術演算

```
def sub(a, b):
  _a = tf.constant(a)
  _b = tf.constant(b)
  return tf.subtract(_a, _b)
```

```
def mul(a, b):
  _a = tf.constant(a)
  _b = tf.constant(b)
  return tf.multiply(_a, _b)
```

```
def div(a, b):
  _a = tf.constant(a)
  _b = tf.constant(b)
  return tf.div(_a, _b)
```

```
def pow(a, b):
  _a = tf.constant(a)
  _b = tf.constant(b)
  return tf.pow(_a, _b)
```

```
with tf.Session() as sess:
  print(sess.run(sub(1, 2)))
  print(sess.run(mul(4, 5)))
  print(sess.run(div(7, 5)))
  print(sess.run(pow(3, 2)))
```
